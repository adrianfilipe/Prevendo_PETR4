```python
import requests
import pandas as pd

# Defina a URL da API
url = "http://servicodados.ibge.gov.br/api/v3/noticias/"

# Parâmetros da requisição (opcionais)
params = {
    'tipo': 'noticia',  # Tipo de dado: 'noticia' ou 'release'
    'de': '2009-01-01',  # Data mínima de publicação no formato AAAA-MM-DD
    'ate': '2017-12-30',  # Data máxima de publicação no formato AAAA-MM-DD
    'busca': 'petrobras',  # Termo de busca (se necessário)
}

# Fazendo a requisição GET
response = requests.get(url, params=params)

# Converte a resposta para JSON
dados = response.json()

# Extraindo a lista de itens (notícias) da resposta
noticias = dados['items']
    
# Convertendo para DataFrame do pandas
df = pd.DataFrame(noticias)
df
```

```python
from GoogleNews import GoogleNews
import pandas as pd

googlenews = GoogleNews(lang='pt-br')
googlenews.set_time_range('01/01/2010','12/31/2023')
googlenews.get_news('petrobras')
googlenews.search('petrobras')

# Inicializa uma lista vazia para armazenar os resultados
all_news = []

# Loop para percorrer as páginas
for i in range(1, 900001):  # Pega as 10 primeiras páginas
    googlenews.get_page(i)
    # Adiciona os resultados da página atual à lista `all_news`
    all_news.extend(googlenews.result())

# Converte a lista para um DataFrame
df = pd.DataFrame(all_news)

# Exibe o DataFrame
df
```
