```python
from GoogleNews import GoogleNews
googlenews = GoogleNews()
```

```python
googlenews = GoogleNews(lang='pt-br')
googlenews.set_time_range('02/01/2018','02/28/2020')
googlenews.get_news('petrobras')
googlenews.search('petrobras')
```

```python
result_0 = googlenews.page_at(1)

desc_1 = googlenews.get_texts()
link_1 = googlenews.get_links()
# date_1 = googlenews.get_dates()  # Se a função estiver disponível

for i in list(range(1, 10)):

    result = googlenews.page_at(i)
    desc = googlenews.get_texts()
    link = googlenews.get_links()
    # date = googlenews.get_dates()  # Se a função estiver disponível

    desc_1 = desc_1 + desc
    link_1 = link_1 + link
    # date_1 = date_1 + date  # Adiciona as datas
```

```python
import pandas as pd

column_names = ["description_text", 'link','data']
df = pd.DataFrame(columns = column_names)

df['description_text'] = desc_1
df['link'] = link_1
# df['data'] = date_1
display(df)

df.to_csv('google_news.csv', index = False)
```
